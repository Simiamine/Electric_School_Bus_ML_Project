```python
from preprocessing import prepare_district_data

# üßº Pipeline complet sur df_districts
df_district_ready = prepare_district_data(df_districts)

# üëÄ Aper√ßu des donn√©es pr√™tes pour le clustering
df_district_ready.head()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# üéØ Appliquer PCA
features = df_district_ready.drop(columns=["locale_type"])  # on exclut la cat√©gorie
pca = PCA(n_components=2)
pca_result = pca.fit_transform(features)

# üì¶ Cr√©er un DataFrame des composantes principales
df_pca = pd.DataFrame(pca_result, columns=["PC1", "PC2"])
df_pca["locale_type"] = df_district_ready["locale_type"]

# üìä Visualisation 2D
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_pca, x="PC1", y="PC2", hue="locale_type", palette="Set2", s=30)
plt.title("PCA des districts (2 composantes principales)")
plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]*100:.1f}% variance)")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]*100:.1f}% variance)")
plt.legend(title="Locale Type")
plt.grid(True)
plt.show()
```

```python
# Courbe de variance expliqu√©e
plt.figure(figsize=(8, 4))
plt.plot(range(1, len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_.cumsum(), marker='o')
plt.title("Variance cumul√©e expliqu√©e par les composantes principales")
plt.xlabel("Nombre de composantes")
plt.ylabel("Variance expliqu√©e cumul√©e")
plt.grid(True)
plt.show()
```

## üî¢ Clustering avec KMeans

Maintenant que nos donn√©es sont pr√™tes et r√©duites en 2 dimensions principales gr√¢ce au PCA, nous allons appliquer un algorithme de clustering.

Nous commen√ßons par le **KMeans**, une m√©thode classique et efficace pour segmenter des entit√©s en groupes homog√®nes selon leur position dans l‚Äôespace vectoriel.  
L‚Äôobjectif est ici de regrouper les districts scolaires selon leurs caract√©ristiques environnementales, sociales et techniques li√©es √† l‚Äôadoption des bus √©lectriques.

Nous allons :
1. Utiliser la **m√©thode du coude** pour d√©terminer un nombre optimal de clusters.
2. Calculer le **silhouette score** pour √©valuer la qualit√© des clusters.
3. Appliquer **KMeans** avec le meilleur `k`.
4. Visualiser les r√©sultats et interpr√©ter les profils.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

inertia = []
silhouette_scores = []
k_values = range(2, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(features, kmeans.labels_))

# üìà Affichage de la m√©thode du coude + silhouette
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(k_values, inertia, marker='o')
plt.title("M√©thode du coude")
plt.xlabel("Nombre de clusters (k)")
plt.ylabel("Inertie (distorsion)")
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(k_values, silhouette_scores, marker='o', color='orange')
plt.title("Score de silhouette")
plt.xlabel("Nombre de clusters (k)")
plt.ylabel("Silhouette score")
plt.grid(True)

plt.tight_layout()
plt.show()
```

### üîç Choix du nombre de clusters

- La **m√©thode du coude** consiste √† rep√©rer un "coude" dans la courbe d'inertie : c‚Äôest √† cet endroit que l‚Äôajout de nouveaux clusters n‚Äôapporte plus beaucoup d‚Äôam√©lioration.
- Le **score de silhouette** indique dans quelle mesure les points sont bien regroup√©s dans leurs clusters respectifs (valeurs proches de 1 = tr√®s bon clustering).

Sur les graphiques ci-dessus, nous choisissons un `k` adapt√© en cherchant un **compromis entre faible inertie et bon score de silhouette**.

```python
# üí° √Ä adapter selon le graphique
k_optimal = 4

# Clustering final
kmeans_final = KMeans(n_clusters=k_optimal, random_state=42)
df_pca["cluster"] = kmeans_final.fit_predict(features)

# üé® Visualisation des clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_pca, x="PC1", y="PC2", hue="cluster", palette="Set2", s=30)
plt.title(f"KMeans clustering (k={k_optimal}) sur PCA")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.legend(title="Cluster")
plt.grid(True)
plt.show()
```
